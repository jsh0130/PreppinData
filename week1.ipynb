```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
original = pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
```

```python
original.head(10)
```

```python
original.info()
```

```python
cont = pd.DataFrame( columns=['Academic Year','Pupil\'s Name','Parental Contact Full Name','Parental Contact Email Address'])
cont
```

```python
cont['Pupil\'s Name'] = original['pupil last name'] + ', ' + original['pupil first name']
cont
#cont['Pupil\'s Name'] = original['pupil last name'].str.cat(original['pupil first name'], sep = ', ')
```

```python
original['Date of Birth'] = pd.to_datetime(original['Date of Birth'])
original
original.info()
```

```python
original.loc[original['Date of Birth'] >= '09/01/2011', 'year'] = 4
original.loc[original['Date of Birth'] >= '09/01/2012', 'year'] = 3
original.loc[original['Date of Birth'] >= '09/01/2013', 'year'] = 2
original.loc[original['Date of Birth'] >= '09/01/2014', 'year'] = 1

original['year'] = original['year'].astype(int)
original
```

```python
cont['Academic Year'] = original['year']
cont
```

```python
if (original['Parental Contact'] ==1).any :
    original['Parental Contact Full Name'] = original['pupil last name'] + ', ' + original['Parental Contact Name_1']
elif (original['Parental Contact'] ==2).any :
    original['Parental Contact Full Name'] = original['pupil last name'] + ', ' + original['Parental Contact Name_2']
original
```

```python
cont['Parental Contact Full Name'] = original['Parental Contact Full Name']
cont
```

```python
cont[['lname','fname']] = cont['Parental Contact Full Name'].str.split(', ', n = 1, expand = True)
cont
```

```python
cont['Parental Contact Email Address'] = cont['fname'] + '.' + cont['lname'] + '@' + original['Preferred Contact Employer'] + '.com'
cont
```

```python
cont = cont.drop(['fname','lname'],axis = 1)
cont
```

```python
cont.to_csv('week1.csv')
```
