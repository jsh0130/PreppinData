```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
df2 = pd.read_csv('PD 2021 WK 1 to 4 ideas - Preferences of Travel.csv')
```

```python
df1
```

```python
df2
```

```python
df = pd.merge(left = df1, right = df2, how = 'inner', left_on = 'id', right_on = 'Student ID')
df
```

```python
pd.unique(df[['M','Tu','W','Th','F']].values.ravel())
```

```python
for i in range(0,1000):
    for j in range(10,15):
        if (df.iloc[i,j].startswith('W')):
            df.iloc[i,j] = 'Walk'
        elif (df.iloc[i,j].startswith('B')) :
            df.iloc[i,j] = 'Bicycle'
        elif (df.iloc[i,j].startswith('C')) :
            df.iloc[i,j] = 'Car'
        elif (df.iloc[i,j].startswith('S')) :
            df.iloc[i,j] = 'Scooter'
        elif (df.iloc[i,j].startswith('He')) :
            df.iloc[i,j] = 'Helicopter'
df
```

```python
pd.unique(df[['M','Tu','W','Th','F']].values.ravel())
```

```python
df = df[['id','M','Tu','W','Th','F']]
df
```

```python
df = df.melt(id_vars=['id'])
df
```

```python
df.columns = ['id','Weekday','Method of Travel']
df
```

```python
ns = ['Aeroplane','Car','Helicopter','Van']
"""
if (df['Method of Travel'].isin(ns)).any:
    df['Sustainable?'] = 'Non-Sustainable'
else : df['Sustainable?'] = 'Sustainable'
"""
df.loc[df['Method of Travel'].isin(ns), 'Sustainable?'] = 'Non-Sustainable'
df.loc[~df['Method of Travel'].isin(ns), 'Sustainable?'] = 'Sustainable'
df
```

```python
df['Number of Trips'] = df.groupby(['Method of Travel','Weekday'])['id'].transform('count')
df
```

```python
df = df.drop(columns='id', axis = 1)
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.reset_index()
df = df.drop(columns='index', axis=1)
df
```

```python
df = df.sort_values(by='Method of Travel', ascending=True)
df
```

```python
df['Trips per day'] = 1000
df
```

```python
df.info()
```

```python
df['% of trips per day'] = (df['Number of Trips']/df['Trips per day']).round(2)
df
```

```python
df = df[['Sustainable?','Method of Travel','Weekday','Number of Trips','Trips per day','% of trips per day']]
df
```

```python
df.to_csv('week4.csv')
```
