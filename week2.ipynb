```python
import pandas as pd
import numpy as np
df = pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
```

```python
df
```

```python
df.info()
```

```python
from datetime import datetime, timedelta
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
df.info()
```

```python
df = df.iloc[:,[1,2,4]]
df
```

```python
df.info()
```

```python
bd = pd.DataFrame()
bd['Pupil Name'] = df['pupil first name'] + ' ' + df['pupil last name']
bd
```

```python
bd['Date of Birth'] = df['Date of Birth']
bd
```

```python
bd['This Year\'s Birthday'] = str(datetime.today().year) + '-' + df['Date of Birth'].dt.strftime("%m-%d")
bd['This Year\'s Birthday'] = pd.to_datetime(bd['This Year\'s Birthday'])
bd.info()
```

```python
bd['Month'] = bd['Date of Birth'].dt.strftime('%B')
bd
```

```python
bd['Cake Needed On'] = bd['This Year\'s Birthday'].dt.strftime('%A')
bd
```

```python
bd = bd.replace({'Cake Needed On':['Saturday','Sunday']},'Friday')
bd.head(8)
```

```python
bd.groupby(['Month','Cake Needed On'])['Pupil Name'].count().unstack()
```

```python
bd['BDs per Weekday and Month'] = bd.groupby(['Month','Cake Needed On'])['Pupil Name'].transform('count')
bd
```

```python
bd.to_csv('week2.csv')
```
