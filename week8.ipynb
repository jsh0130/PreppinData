```python
import pandas as pd
import numpy as np

stat = pd.read_excel('input_pkmn_stats_and_evolutions.xlsx', sheet_name = 'pkmn_stats')
evo = pd.read_excel('input_pkmn_stats_and_evolutions.xlsx', sheet_name = 'pkmn_evolutions')
```

```python
stat
```

```python
evo
```

```python
stat = stat.drop(columns = ['height', 'weight', 'evolves_from'])
stat
```

```python
stat2 = pd.melt(stat, id_vars = ['name','pokedex_number','gen_introduced'],var_name = 'combat_factors', value_name = 'values')
stat2
```

```python
stat2 = stat2.sort_values(by='pokedex_number')
stat2
```

```python
stat2 =stat2.reset_index(drop=True)
stat2
```

```python
stat3 = stat2.groupby(['name','pokedex_number','gen_introduced']).sum()
stat3
```

```python
stat3 = stat3.sort_values(by = 'pokedex_number')
stat3
```

```python
stat3.reset_index(inplace = True)
stat3
```

```python
evo
```

```python
df = pd.merge(evo, stat3, left_on = 'Stage_1', right_on = 'name', how='left')
df
```

```python
df.rename(columns={'values':'initial_combat_power'}, inplace = True)
df
```

```python
df = df.drop(columns='name')
df
```

```python
df = df.dropna(subset=['Stage_2'])
df
```

```python
fin = stat3[['name','values']]
fin
```

```python
df = pd.merge(df, fin, left_on='Stage_2', right_on='name', how='left')
df
```

```python
df = pd.merge(df, fin, left_on='Stage_3', right_on='name', how='left')
df
```

```python
df['final_combat_power'] = df['values_y']
df.loc[pd.isnull(df['final_combat_power']),'final_combat_power']=df['values_x']
df
```

```python
df = df.drop(columns=['name_x','values_x','name_y','values_y'])
df
```

```python
df.info()
```

```python
df['final_combat_power'] = df['final_combat_power'].astype(int)
df.info()
```

```python
df['combat_power_increase'] = df['final_combat_power']/df['initial_combat_power']-1
df
```

```python
df.to_csv('Output22_8.csv')
```
