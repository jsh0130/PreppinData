```python
import pandas as pd
import numpy as np

people = pd.read_excel('PeopleData.xlsx', sheet_name = 'People')
location = pd.read_excel('PeopleData.xlsx', sheet_name = 'Location')
leader = pd.read_excel('PeopleData.xlsx', sheet_name = 'Leaders')
date = pd.read_excel('PeopleData.xlsx', sheet_name = 'Date Dim')
goal = pd.read_excel('PeopleData.xlsx', sheet_name = 'Goals')
```

```python
df1 = pd.merge(people, leader, left_on = 'Leader 1', right_on = 'id', how = 'inner')
df1
```

```python
df1['Agent Name'] = df1['last_name_x'] + ', ' + df1['first_name_x']
df1
```

```python
df1['Leader Name'] = df1['last_name_y'] + ', ' + df1['first_name_y']
df1
```

```python
df1 = df1.drop(columns = ['first_name_x','first_name_y','last_name_x','last_name_y'], axis = 1)
df1
```

```python
df1 = pd.merge(df1, location , on = 'Location ID' , how = 'inner')
df1
```

```python
df1 = df1.rename(columns={'id_x':'id'})
df1 = df1.drop(columns = 'id_y', axis=1)
df1
```

```python
date.info()
```

```python
date = date.loc[date['Month Start Date']<'2022-01-01']
date
```

```python
df1['key'] = 1
date['key'] = 1
df1 = pd.merge(df1, date, on = 'key', how = 'outer')
df1
```

```python
del df1['key']
df1
```

```python
df2 = pd.read_excel('MetricData2021.xlsx', sheet_name= None)
df2
```

```python
i = 1
for key in df2.keys():
    df2[key]['month'] = '2021-{}-01'.format(i)
    i += 1
df2
```

```python
df2['Dec'].info()
```

```python
df2 = pd.concat(df2, ignore_index = True)
df2
```

```python
df2.info()
```

```python
df3 = df2.copy()
df3
```

```python
df3 = df3.replace(np.NaN, 0)
df3
```

```python
df3['Calls Answered'] = df3['Calls Answered'] + df3['Answered']
df3['Calls Offered'] = df3['Calls Offered'] + df3['Offered']
df3['Calls Not Answered'] = df3['Calls Not Answered'] + df3['Not Answered']
df3
```

```python
df3 = df3.drop(columns= ['Offered', 'Not Answered', 'Answered'], axis=1)
df3
```

```python
df3['month'] = pd.to_datetime(df3['month'])
df3
```

```python
df3.info()
```

```python
df = pd.merge(df1, df3, left_on = ['Month Start Date','id'], right_on = ['month','AgentID'], how = 'outer')
df
```

```python
df= df[['id','Agent Name','Leader 1','Leader Name','Month Start Date','Location','Calls Answered','Calls Not Answered','Calls Offered','Total Duration','Transfers','Sentiment']]
```

```python
df
```

```python
df.insert(8, 'Not Answered Rate', df['Calls Not Answered']/(df['Calls Offered']))
df
```

```python
df.insert(9, 'Met Not Answered Rate', df['Not Answered Rate']<5/100)
df
```

```python
df.insert(10, 'Not Answered Percent < 5', 5)
df
```

```python
df.insert(13, 'Agent Avg Duration', df['Total Duration']/(df['Calls Offered']))
df
```

```python
df['Sentiment Score >= 0'] =0
df
```

```python
df['Met Sentiment Goal'] = (df['Sentiment']>=0)
df
```
