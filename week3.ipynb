```python
import pandas as pd
w1 = pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
w3 = pd.read_csv('PD 2022 WK 3 Grades.csv')
```

```python
w1
```

```python
w3
```

```python
df = pd.merge(left = w1, right = w3, how = 'inner', left_on = 'id', right_on = 'Student ID')
df
```

```python
df.info()
```

```python
df = df.drop(df.columns[[5,6,7,8]], axis = 1)
```

```python
df.info()
```

```python
df2 = df.iloc[:,6:13]
df2
```

```python
df['Student\'s Avg Score'] = df2.mean(axis=1).round(1)
df
```

```python
df.info()
```

```python
df = df.drop(df.columns[[0,1,2,4]], axis = 1)
```

```python
df
```

```python
df.iloc[0,0]
```

```python
for i in range(0,1000):
    cnt = 0
    for j in range(2,9):
        if df.iloc[i,j] >= 75:
            cnt += 1
    df.loc[i, 'Passed Subjects'] = cnt
df
```

```python
df.info()
```

```python
df['Passed Subjects'] = df['Passed Subjects'].astype(int)
df.info()
```

```python
done = df[['Passed Subjects','Student\'s Avg Score', 'Student ID', 'gender']]
done
```

```python
done.rename(columns={'gender':'GENDER'}, inplace = True)
done
```

```python
done.to_csv('week3.csv')
```
